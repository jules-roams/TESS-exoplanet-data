# The purpose of this notebook is to produce preliminary plots to get an idea of the feature space and apply some basic analysis techniques

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('exoplanets_table.csv')
data
```

## Make some preliminary plots

```python
plt.figure(figsize = [10,8])
transit_dur = data['Transit_Duration']
period = data['Period']
plt.ylim(-20,1250)
plt.xlim(-2,30)
plt.xlabel('Transit Duration [hr]', fontsize = 14)
plt.ylabel('Orbital Period [days]', fontsize = 14)
plt.plot(transit_dur, period, 'm*')
```
